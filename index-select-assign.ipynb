# Indexing, Selecting & Assigning

````python
import pandas as pd
reviews = pd.read_csv("./data/wine-reviews/winemag-data-130k-v2.csv", index_col=0)
pd.set_option('display.max_rows', 5)
````

### Native Accessors

Pandas carries over the ways of indexing data found in native Python objects. So we can select a specific Series from a DataFrame as follows:

````python
reviews.country
````

````python
reviews['country']
````

We can drill down to a specific value in a familiar way:

````python
reviews['country'][0]
````

### Indexing in pandas

`pandas` also has its own accessor operators `loc` and `iloc`. These are preferred for more advanced operations. They are both row-first, column-second which is the reverse of native Python.

There are two paradigmns of pandas indexing - `index-based selection` and `label-based selection`.

#### Index-Based Selection

Index-based selection selects data based on its numerical position in the data. `iloc` follows this paradigm.

To select the first row of data in a DataFrame:

````python
reviews.iloc[0]
````

To select a column:

````python
reviews.iloc[:, 0]
````

The `:` operator by itself means 'everything'. When combined with other selectors, it can be used to indicate a range of values.

For example, to select the country column from only the 1st, 2nd and 3rd row:

````python
reviews.iloc[:3, 0]
````

Here are some related examples.

Select only the 2nd and 3rd entries:

````python
reviews.iloc[1:3, 0]
````

Use a list to select the entries:

````python
reviews.iloc[[0,1,2], 0]
````

Use a negative number to select starting from the end of the entries:

````python
reviews.iloc[-5:]
````

#### Label-Based Selection

In label-based selection, the data index value is used to select data. This paradigm is followed by the `loc` operator.

To get the first entry in reviews:

````python
reviews.loc[0, 'country']
````

Since `loc` uses information in the dataset's indices, it can be simpler to use:

````python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
````

#### Note on loc vs iloc

One difference to note is that the paradigms use slightly different indexing schemes.

`iloc` excludes the last element of a range so
```
0:10
```
will return `0, ..., 9`.

`loc` includes the last element of a range so
```
0:10
```
will return `0, ..., 10`.

This difference is due to fact that `loc` can index other data types such as strings, where this inclusive indexing would make more sense.

### Manipulating the Index

Label-based selection gets its power from the labels in the index and we can manipulate this index however we see fit.

For example, we can change the index to be the title column:

````python
reviews.set_index('title')
````

### Conditional Selection

To check if each wine is italian we could produce a Series of booleans based on the country of each record:

````python
reviews.country == 'Italy'
````

We select the relevant rows from the dataset by using this inside a `loc`:

````python
reviews.loc[reviews.country == 'Italy']
````

If we wanted to know which wines are italian AND also better than average (in our case at least 90 points) we can combine statements:

````python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
````

`pandas` includes a few built-in conditional selectors. Here are two examples.

We can select wines only from Italy or France using `isin`:

````python
reviews.loc[reviews.country.isin(['Italy', 'France'])]
````

We can filter out wines without a price tag using `isnull` or `notnull`:

````python
reviews.loc[reviews.price.notnull()]
````

### Assigning Data

We can easily assign data to a DataFrame. 

To assign a constant value:

````python
reviews['critic'] = 'everyone'

reviews['critic']
````

To assign an iterable of values:

````python
reviews['index_backwards'] = range(len(reviews), 0, -1)
reviews['index_backwards']
````
